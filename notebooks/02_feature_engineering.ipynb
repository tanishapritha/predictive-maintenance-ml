# 02 - Feature Engineering


```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('../data/ai4i2020.csv')
```

## Encode Type and Engineer Features


```python
df['Type'] = df['Type'].map({'L': 0, 'M': 1, 'H': 2})

df['temp_diff'] = df['Process temperature [K]'] - df['Air temperature [K]']
df['power'] = df['Torque [Nm]'] * df['Rotational speed [rpm]']
df['wear_rate'] = df['Tool wear [min]'] / df['Rotational speed [rpm]']

max_wear = df['Tool wear [min]'].max()
df['RUL'] = max_wear - df['Tool wear [min]']

# Strip brackets for model compatibility
df.columns = [col.replace('[', '').replace(']', '').replace('<', '').strip() for col in df.columns]
```

## Scaling and Saving


```python
scaler = StandardScaler()
features = ['Air temperature K', 'Process temperature K', 'Rotational speed rpm', 'Torque Nm', 'Tool wear min', 'temp_diff', 'power', 'wear_rate']
df[features] = scaler.fit_transform(df[features])

df.to_csv('../data/processed.csv', index=False)
```
