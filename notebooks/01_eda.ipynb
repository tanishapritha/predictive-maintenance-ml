# 01 - Exploratory Data Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/ai4i2020.csv')
df.head()
```

## Dataset Shape, Dtypes, Null Check


```python
print('Shape:', df.shape)
print('Nulls:', df.isnull().sum())
print(df.dtypes)
```

## Visualizations


```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Interpretation
- **Tool wear**: As tool wear approaches max capacity, failure probability increases non-linearly.
- **Torque-Stress**: High torque under high rotational speeds introduces immense shear stress, pushing equipment closer to threshold limits.
- **Temperature Delta**: The difference between Process and Air temperatures represents the heat generated by friction and energy loss. Sudden spikes denote imminent overheating faults.
